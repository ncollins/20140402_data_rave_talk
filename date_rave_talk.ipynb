```python
%matplotlib inline

import csv
import itertools
import datetime

import dateutil
import pandas
import math, numpy
import matplotlib.pyplot as plt

from whoosh import index, fields, qparser
import seaborn
```

```python
# csv details

csv_headings = ['congress', 'chamber', 'number', 'date', 'id', 'title', 'path']

# utility functions

def convert(k, v):
    if k == 'date':
        return k, dateutil.parser.parse(v)
    elif k == 'congress' or k == 'number':
        return k, int(v)
    else:
        return k, unicode(v)

def group_by_month(items):
    items = sorted(items, key=lambda x: x['date'])
    print(type(items))
    for i in items:
        i['month'] = datetime.date(i['date'].year, i['date'].month, 1)
    return itertools.groupby(items, key=lambda x: x['month'])
```

```python
# build index

schema = fields.Schema(congress=fields.NUMERIC(stored=True),
                       chamber=fields.TEXT(stored=True),
                       number=fields.NUMERIC(stored=True),
                       date=fields.DATETIME(stored=True),
                       week=fields.NUMERIC(stored=True),
                       id=fields.TEXT(stored=True),
                       title=fields.TEXT(stored=True),
                       path=fields.ID(stored=True),
                       length=fields.NUMERIC(stored=True),
                       content=fields.TEXT)

ix = index.create_in("index", schema)
```

```python
# create writer
writer = ix.writer()
```

```python
# get metadata from csv file

with open('bills.csv', 'r') as csv_file:
    reader = csv.DictReader(csv_file, delimiter=',', quotechar='|', fieldnames=csv_headings)
    rows = []
    for row in reader:
        with open(row['path'], 'r') as text_file:
            content = text_file.read()
            r = dict(convert(k,v) for k,v in row.items())
            r['week'] = r['date'].timetuple().tm_yday % 52
            r['length'] = len(content.split())
            rows.append(r)
            writer.add_document(content=unicode(content), **r)

table = pandas.DataFrame(rows)
writer.commit()

print len(table)
```

```python
# create Searcher
# >>> searcher = ix.searcher()
# >>> searcher.close()

with ix.searcher() as s:
    qp = qparser.QueryParser('content', schema=ix.schema)
    q_tax = qp.parse(u'(title:debt)^2 OR tax')
    results = s.search(q_tax, limit=None)
    
    print '#results = {0}\n'.format(len(results))
    print results[:2]
    print
    
    search_table = pandas.DataFrame(x.fields() for x in results)
    
    #hit = results[0]
    #more = hit.more_like_this("title")
    #for m in more:
    #    print m['title']
    
print
print search_table[:2]
```
